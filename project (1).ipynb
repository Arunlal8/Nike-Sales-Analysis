```python
import pandas as pd 
```

```python
df = pd.read_csv(r'C:\Users\arun8\Downloads\Nike_Sales_Uncleaned.csv')
```

```python
print(df.dtypes)
```

```python
df.head()
df.info()
df.describe(include='all')
```

```python
df.isnull().sum()
df.fillna({'region': 'Unknown'}, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
```

```python
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['discount_rate'] = df['discount_applied'] / 100
```

```python
df = df[df['revenue'] > 0]
```

```python
df['region'].value_counts()
df['product_line'].value_counts()
```

```python
print(df.columns.tolist())
```

```python
df.groupby('region')['revenue'].sum()
df.groupby(['region', 'product_line'])['revenue'].mean()
```

```python
df.pivot_table(index='region', columns='month', values='revenue', aggfunc='sum')
df.corr(numeric_only=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(x='region', y='revenue', data=df)
plt.title('Sales by Region')
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_sales = df.groupby('month')['revenue'].sum()
monthly_sales.plot(kind='line', marker='o', title='Monthly Sales Trend')
plt.show()
```

```python
df['product_line'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Product Category Distribution')
plt.ylabel('')
plt.show()
```

```python
sns.histplot(df['discount_applied'], bins=20, kde=True)
plt.title('Discount Distribution')
plt.show()
```

```python
sns.boxplot(x='product_line', y='revenue', data=df)
plt.title('Sales Distribution by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x='discount_applied', y='revenue', data=df)
plt.title('Sales vs Discount')
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
pivot = df.pivot_table(index='region', columns='month', values='revenue', aggfunc='sum')
sns.heatmap(pivot, cmap='YlGnBu', ax=ax)
plt.title('Monthly Sales by Region')
plt.show()
```

```python

#KEY FINDINGS

1. Footwear is the top-selling category

.Footwear generated the highest sales and contributed the largest share of revenue.

.This shows Nike customers prefer footwear over apparel and accessories.

2. Sales show clear seasonal trends

.Certain months, especially November and December, had peak sales.

.This indicates festive seasons, offers, or holiday shopping increases sales.

3. Apparel and Accessories underperform

.These categories have lower demand and lower sales volume compared to footwear.

.They may need better promotion, pricing strategies, or new product designs.

4. Sales quantity and revenue are strongly related

.Products with high quantity sold also generate high revenue.

.This means Nikes popular products directly drive overall sales growth.

5. Some months show low sales

.Months like February and March show dips in sales.

.These can be off-season periods or months with fewer promotions.

6. Consistent growth in certain product lines

.Some product types show steady growth month-by-month.

.These can be strong areas for future investment.
```

```python

# KEY INSIGHTS

1. Focus more on Footwear

.Since footwear is the best performer, Nike should increase marketing, offers, and product launches in this category.

2. Seasonal planning is important

.Nike should prepare inventory and marketing campaigns before peak seasons, especially during holidays.

3. Improve low-performing categories

.Apparel and accessories need better marketing, discounts, or updated designs to boost sales.

4. Use data to plan promotions

.Months with low sales can benefit from campaigns, discounts, or special events to increase customer interest.

5. Product variety matters

.Popular products should be expanded with new styles or variations since they drive the highest revenue.

6. Data analytics helps decision making

.Managers can use this analysis to understand customer demand and plan stock, marketing, and pricing strategies more effectively.
```
